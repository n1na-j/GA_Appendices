```python
import pandas as pd

df = pd.read_excel(r"C:\Users\ninaj\Documents\Data-Driven Design Y2\Graduation Project\Week 9\data\ah_data.xlsx")

df.head()
```

```python
# Add new column which categorizes the price into "Low budget" (0-1.50 euros), "Medium budget" (1.50-3.00 euros) and "High budget" (More than 3.00 euros)
df['Price_cat'] = pd.cut(df['Price'], bins = [0, 1.50, 3.00, 10], include_lowest=True, labels= ['Low budget', 'Medium budget', 'High budget'])

# If Title contians the word "Glutenfree", add True/False to column "Glutenfree"
glutenfree = ['Glutenvrij', 'glutenvrij']
df['Glutenfree'] = df['Description'].str.contains('|'.join(glutenfree))

# If Description contains the word "Skal", add True/False to column "BIO"
df['BIO'] = df['Description'].str.contains("Skal")

# If Description contains the word "Fairtrade", add True/False to column "Fairtrade"
df['Fairtrade'] = df['Description'].str.contains("Fairtrade")

# if Description contains the word "UTZ", add True/False to column "UTZ"
df['UTZ'] = df['Description'].str.contains("UTZ")

# if Description contains the word "Vegan", add True/False to column "Vegan"
vegan = ['Vegan', 'vegan']
df['Vegan'] = df['Description'].str.contains('|'.join(vegan))

# if Description contains the word "Vegatarisch", add True/False to column "Vegetarian"
vegetarian = ['Vegetarisch', 'vegetarisch']
df['Vegetarian'] = df['Description'].str.contains('|'.join(vegetarian))

# if Description contains the word "Low in sugar", "Laag in suiker", "Zoetstoffen", "Zonder toegevoegde suikers", add True/False to column "laag in suikers"
low_sugar = ['Laag in suiker', 'suikervrij', 'suikervrije', 'Suikervrij', 'Suikervrije', 'low in sugar', 'Zonder toegevoegde suikers', 'zoetstoffen']
df['Low_sugar'] = df['Description'].str.contains('|'.join(low_sugar)) 

# if Description contains the word "Low in sugar", "Laag in suiker", "Zoetstoffen", "Zonder toegevoegde suikers", add True/False to column "laag in suikers"
low_fat = ['Vetgereduceerd', 'Vetvrij', 'vetvrij', 'vetvrije', 'Vetvrije', 'vetarm', 'Vetarm', 'laag in vet', 'Laag in vet']
df['Low_fat'] = df['Description'].str.contains('|'.join(low_fat)) 

# if Description contains the word "Low in sugar", "Laag in suiker", "Zoetstoffen", "Zonder toegevoegde suikers", add True/False to column "laag in suikers"
plant_based = ['Plantaardig', 'plantaardig', 'Plantaardige', 'plantaardige', 'soja']
df['Plant_based'] = df['Description'].str.contains('|'.join(plant_based)) 

```

```python
# Transform "True" and "False" to 1 and 0
df['Glutenfree'] = pd.to_numeric(df['Glutenfree'], errors='coerce').convert_dtypes()
df['BIO'] = pd.to_numeric(df['BIO'], errors='coerce').convert_dtypes()
df['Fairtrade'] = pd.to_numeric(df['Fairtrade'], errors='coerce').convert_dtypes()
df['UTZ'] = pd.to_numeric(df['UTZ'], errors='coerce').convert_dtypes()
df['Vegan'] = pd.to_numeric(df['Vegan'], errors='coerce').convert_dtypes()
df['Vegetarian'] = pd.to_numeric(df['Vegetarian'], errors='coerce').convert_dtypes()
df['Low_sugar'] = pd.to_numeric(df['Low_sugar'], errors='coerce').convert_dtypes()
df['Low_fat'] = pd.to_numeric(df['Low_fat'], errors='coerce').convert_dtypes()
df['Plant_based'] = pd.to_numeric(df['Plant_based'], errors='coerce').convert_dtypes()

# Checks
print(df[['Price','Price_cat']])
print("Glutenfree check", df.iloc[599])
print("BIO check", df.iloc[21])
print("Fairtrade check", df.iloc[1668])
print("UTZ check", df.iloc[2278])
print("Vegan check", df.iloc[750])
print("Vegetarian check", df.iloc[853])
print("Low sugar check", df.iloc[1499])
print("Low fat check", df.iloc[1470])
print("Plant based check", df.iloc[1459])
```
